```python
import nltk
from textblob import TextBlob
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import brown

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')

tokenizer = RegexpTokenizer(r'\w+')
```

```python
stop_words = stopwords.words('english')
```

```python
text = '/content/drive/MyDrive/CS6678Homework6/Shakespeare.txt'
file = open(text).read()
shakespeare = TextBlob(file)
shakespeare_tokenized = tokenizer.tokenize(file)
```

```python
shakespeare_tokenized
```

```python
words = set(w.lower() for w in brown.words())
```

```python
words_dictionary = []
```

```python
for s in shakespeare_tokenized:
    if s in words:
        words_dictionary.append(s)
```

```python
words_dictionary
```

```python
all_lower_words_dictionary = [w.lower() for w in words_dictionary if w.isalpha()]
```

```python
len(set(all_lower_words_dictionary))
```

```python
sorted(set(all_lower_words_dictionary))
```
