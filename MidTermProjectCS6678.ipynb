```python
covid_death_covid_death_dataset = pd.read_csv('/content/drive/MyDrive/Covid/COVID19inAlabama.csv')
```

```python
covid_death_dataset = covid_death_covid_death_dataset.drop(labels=[0,68], axis =0)
covid_death_dataset['Deaths'] = pd.to_numeric(covid_death_dataset['Deaths'])
covid_death_dataset['Cases'] = pd.to_numeric(covid_death_dataset['Cases'])
covid_death_dataset['Deaths per Case'] = covid_death_dataset['Deaths']/covid_death_dataset['Cases']
```

```python
covid_death_dataset
```

```python
covid_death_dataset['Death Percentage'] = covid_death_dataset['Deaths per Case']*100
```

```python
covid_death_dataset.sort_values(by=['Deaths per Case'], ascending = False)
```

```python
color = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
plt.figure(figsize=(10,6))
plt.title("COVID-19 Death Percentage by County")
covid_death_plot = plt.bar(covid_death_dataset['Counties'], covid_death_dataset['Death Percentage'], color = color)

plt.xticks(rotation=90)
plt.tight_layout()
```
